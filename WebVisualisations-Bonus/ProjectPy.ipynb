## Generate Health Data List

```python
# Dependencies
import os
import csv
import requests
import pandas as pd
import json
import matplotlib.pyplot as plt
#from config import weather_api_key
from pprint import pprint

import scipy.stats as st
from scipy.stats import linregress
import numpy as np
#from scipy.stats import sem
```

```python
# Set up limits to retrieve sample of cities from World Cities csv, by Latitude grouping
# Will be using World Cities listing from https://simplemaps.com/data/world-cities
# and my allocation of latitude groupings to spread the cities across the latitudes.
# Set up lists to store retrieved data from Open Weather
Time_period=[]
Week_no=[]
Year=[]
Range_deaths_2015_2019=[]
Covid_infections=[]
Avg_deaths_2015_2019=[]
Deaths_2020_2021=[]
Time_point=[]

# Open the CSV
# Set path for file
csvpath = os.path.join("Resources","Doctor certified deaths and COVID-19 infections Australia vs benchmarks.csv")
print(csvpath)
with open(csvpath) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    csv_header = next(csvreader)
    print(f"CSV Header: {csv_header}")
```

```python
# Loop through looking for the Cities for their latitude grouping and store them up to their maximum allowed.
with open(csvpath) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    csv_header = next(csvreader)
    csvreader = csv.reader(csvfile, delimiter=",")
#    print(f"CSV Header: {csv_header}")   
    cnt=0
    for row in csvreader:
        Time_period.append(row[0])
        Week_no.append(row[1])
        Year.append(row[2])
        Range_deaths_2015_2019.append(row[3])
        Covid_infections.append(row[4])
        Avg_deaths_2015_2019.append(row[5])
        Deaths_2020_2021.append(row[6])
        Time_point.append(row[7])

```

### Convert Raw Data to DataFrame
* Export the health data into a .csv.
* Display the DataFrame

```python
output_data =zip(Time_period,Week_no,Year,Range_deaths_2015_2019,Covid_infections,Avg_deaths_2015_2019,Deaths_2020_2021,Time_point)
output_data_df = pd.DataFrame(output_data, columns=("Time_period","Week_no","Year","Range_deaths_2015_2019","Covid_infections","Avg_deaths_2015_2019","Deaths_2020_2021","Time_point"))
output_data_html= '"""\n'+output_data_df.to_html(columns=["Time_period","Week_no","Year","Range_deaths_2015_2019","Covid_infections","Avg_deaths_2015_2019","Deaths_2020_2021","Time_point"],index=False)+'\n"""'
print(output_data_html)
```

```python
with open("Resources/healthdata.html","w") as file:
    file.write(output_data_html)
    file.close()
```
